```python
# Importing the all important libraries and modules
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
#defining train and test data
(X_train, y_train),(X_test, y_test)=keras.datasets.mnist.load_data()
```

```python
#scaling the complete data to make it between 0 to 1 because it is good for accuracy
X_train=X_train/255
X_test=X_test/255
```

```python
X_train[0]
```

```python
# Flattened the data to make it in one dimension tensors
X_train_flattened = X_train.reshape(len(X_train),28*28)
X_test_flattened = X_test.reshape(len(X_test),28*28)
```

```python
print(X_train_flattened.shape)
print(X_test_flattened.shape)
```

```python
X_train_flattened[0]
```

```python
# Training the model
model=keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,),activation='sigmoid')
])
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
plt.matshow(X_test[5])
```

```python
y_predicted=model.predict(X_test_flattened)
```

```python
y_predicted[5]
```

```python
np.argmax(y_predicted[5])
```

```python
y_predicted_labels=[np.argmax(i) for i in y_predicted]
```

```python
print(y_predicted_labels[:5])
print(y_test[:5])
```

cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
# Training the model with including a hidden layer
model=keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,),activation='relu'),
    keras.layers.Dense(10, input_shape=(784,),activation='sigmoid')
])
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened, y_test)
```
